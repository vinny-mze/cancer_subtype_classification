```python
from imutils import paths
dataset_directory = list(paths.list_images('/kaggle/input/UBC-OCEAN/train_images'))
```

```python
df = pd.read_csv('/kaggle/input/UBC-OCEAN/train.csv')
df.head
```

```python
import os
import sys
import shutil
from pathlib import Path

#creating train step directory
train_dir = '/kaggle/input/UBC-OCEAN/train_images'
Path('/kaggle/working/train_data').mkdir(parents=True, exist_ok=True)
output_dir = '/kaggle/working/train_data'

for image_id, label, _, _, _ in df.values:
    # Create subdirectory with class name
    class_dir = os.path.join(output_dir, str(label))
    if not os.path.exists(class_dir):
        os.makedirs(class_dir)

    src_path = os.path.join(train_dir, f'{image_id}.png')
    dst_path = os.path.join(class_dir, f'{image_id}.png')

    try:
        shutil.copy(src_path, dst_path)
        print(f'Successfully copied {image_id}.png to {class_dir}')
    except IOError as e:
        print(f'Unable to copy file: {e}')
    except Exception as e:
        print(f'Error: {e}')
```

```python
train_directory = '/kaggle/working/train_data'
CLASS_NAMES=['EC','CC','LGSC','HGSC']

CONFIGURATION= {
    'BATCH_SIZE':32,
    'IM_SIZE':256,
    'LEARNING_RATE': 0.001,
    'N_EPOCHS' : 20,
    'DROPOUT_RATE' : 0.0,
    'REGULARIZATION_RATE': 0.0,
    'N_FILTERS' : 6,
    'KERNEL_SIZE': 3,
    'N_STRIDES': 1,
    'POOL_SIZE': 2,
    'N_DENSE_1' : 100,
    'N_DENSE_2': 10,
    'NUM_CLASSES': 4
}
```

```python
train_dataset=tf.keras.utils.image_dataset_from_directory(
    train_directory,
    labels='inferred',
    label_mode='categorical',
    class_names=CLASS_NAMES,
    color_mode='rgb',
    batch_size = CONFIGURATION['BATCH_SIZE'],
    image_size=(CONFIGURATION['IM_SIZE'],CONFIGURATION['IM_SIZE']),
    shuffle=True,
    seed=99
)
```
